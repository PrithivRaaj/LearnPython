# While Loop

```python
count = 0
while (count < 3):
    count += 1
    print("Hello ")
```

```python
count = 0
while (count < 3):
    count = count + 1
    print("Hello Geek")
else:
    print("In Else Block")
```

# Patterns in for loop

```python
n=int(input())
for i in range(n+1):
  for j in range(i):
    print('#',end=' ')
  print()
```

```python
n=int(input())
for i in range(n+1):
  for j in range(n-i):
    print('#',end=' ')
  print()
```

```python
n = int(input())
for i in range(n+1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(i):
        print('#', end=' ')
    print()
```

```python
n = int(input())
for i in range(n, 0, -1):
    for j in range(n-i):
        print(' ', end=' ')
    for j in range(i):
        print('#', end=' ')
    print()
```

```python
n=int(input())
for i in range(n + 1):
        # Print leading spaces
        for j in range(n - i):
            print(" ", end="")

        # Print asterisks for the current row
        for k in range(1, 2*i):
            print("*", end="")
        print()
```

```python
n=int(input())
for i in range(n,0,-1):
        # Print leading spaces
        for j in range(n - i):
            print(" ", end="")

        # Print asterisks for the current row
        for k in range(1, 2*i):
            print("*", end="")
        print()
```

# Pass Statement
To avoid syntax error

```python
i=1
while i<=10:
  print(i)
  if i==5:
  i+=1
```

```python
i=1
while i<=10:
  print(i)
  if i==5:
    pass
  i+=1
```

# Break Statement
break the flow of code when condition hits

```python
i=1
while i<=10:
  print(i)
  if i==5:
    break
  i+=1
```

# Continue Statement
It skips the particular iteration

```python
for i in range(1,11):
  if i==5:
    continue
  print(i)
```
